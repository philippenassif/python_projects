# College majors analysis

In this project we will use many scatter plots and bar charts to see the correlations between certains metrics related to college majors, such as unemployment rate and gender.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna(axis = 0)
```

```python
recent_grads.describe()
```

```python
cleaned_data_count = recent_grads['Rank'].count()
cleaned_data_count
```

Lets start by doing some scatter plots to see correlations between our data columns.

```python
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter', title = 'Median by Sample size', figsize = (7,5))
```

```python
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter', title =  'Unemployment rate by Sample size', figsize = (7,5))
```

```python
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter', title =  'Median by Full_time', figsize = (7,5))
```

```python
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter', title =  'Unemployment rate by ShareWomen', figsize = (7,5))
```

```python
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter', title = 'Median by Men', figsize = (7,5))
```

```python
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter', title = 'Median by Women', figsize = (7,5))
```

Now lets do some histogram plots.

```python
recent_grads['Sample_size'].plot(kind = 'hist', title = 'Sample_size', figsize = (3,3))
```

```python
recent_grads['Sample_size'].hist(bins = 20, range = (0,1000), figsize = (3,3))
```

```python
recent_grads['Median'].plot(kind = 'hist', title = 'Sample_size', figsize = (3,3), rot = 90)
```

```python
import matplotlib.pyplot as plt
cols = ['Employed','Full_time', 'ShareWomen','Unemployment_rate','Men','Women']

fig, ax = plt.subplots()

for i in range(1,6) :
    ax = fig.add_subplot(6,1,i)
    ax = recent_grads[cols[i]].plot(kind = 'hist', rot = 90, title = cols[i])
    plt.show() 
    
#I can't figure out why the first plot is squashed.    
```
