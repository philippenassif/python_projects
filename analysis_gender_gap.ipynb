# Graduate studies gender gap

In this project we want to analyze the gender gaps depending on college degrees. We would like to see all degrees in one big group of charts just to see the trends without too many details.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(14, 18))

## Generate first column of line charts. STEM degrees.
for sp in range(0,17) :
    ax = fig.add_subplot(6,3,sp+1)
    ax.set_yticks([0,100])
    ax.axhline(50, c='lightblue', alpha=0.3)
    if sp == 16 :
        ax = fig.add_subplot(6,3,18)    
    count = int(sp/3)
    plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.5, wspace=0.3)
    if sp in [0,3,6,9,12,15] :
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[count]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[count]], c=cb_orange, label='Men', linewidth=3)
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(stem_cats[count])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
        ax.set_yticks([0,100])

        if sp == 0:
            ax.text(2005, 87, 'Men')
            ax.text(2002, 8, 'Women')
        elif sp == 15:
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(labelbottom='on')
                  
## Generate second column of line charts.           
    elif sp in [1,4,7,10,13] :
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[count]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[count]], c=cb_orange, label='Men', linewidth=3)       
        for key,spine in ax.spines.items():
            spine.set_visible(False)   
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(lib_arts_cats[count])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')

        if sp == 1:
            ax.text(2005, 87, 'Women')
            ax.text(2002, 8, 'Men')
        elif sp == 13:
            ax.tick_params(labelbottom='on')

## Generate third column of line charts.
    else : 
        ax.plot(women_degrees['Year'], women_degrees[other_cats[count]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[count]], c=cb_orange, label='Men', linewidth=3)
        for key,spine in ax.spines.items():
            spine.set_visible(False)  
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(other_cats[count])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')

        if sp == 2:
            ax.text(2005, 87, 'Women')
            ax.text(2002, 8, 'Men')
        elif sp == 16:
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(labelbottom='on')

#we want to save this figure in a png file            
plt.savefig('gender_degrees.png')            
plt.show()
matplotlib.get_backend()
```
